# Purpose:
Plot a user defined function and get the slop of two points on the line graph

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class delta_method():
    '''The delta method is used to approximate the slope of two points on a curve.
    
    Fit the model with two x & y pairs of your choosing using the fit_slope method.
    Plot your function along with the slope of the fitted points.
    '''
    
    def __init__(self):
        self.x = np.linspace(-10,10,1000) # change this if you would like
        
    def fit_slope(self,x2, y2, x1, y1):
        '''fit_slope method:
        computes the slope of two different x & y values
        Returns slope
        '''
        self.x2 = x2
        self.x1 = x1
        self.y2 = y2
        self.y1 = y1
        self.slope = (self.y2-self.y1)/(self.x2-self.x1)
        return self.slope
    
    def plot(self, y_fun):
        '''plot method:
        y_fun: the y values from your self created function.
        The x values used in your function should be the same as
        the x values initialized here, x = np.linspace(-10,10,1000).
        You can change those x values in the code to suit your needs.
        
        Returns: the plots of your function and the fitted slope. 
        '''
        b = self.y2 - self.slope*self.x2
        y3 = self.slope*self.x + b
        plt.plot(self.x, y_fun, label='Function')
        plt.scatter(self.x2,self.y2, c='orange')
        plt.scatter(self.x1,self.y1, c='green')
        plt.plot(self.x,y3, label=f'Slope {self.slope:.2f}')
        plt.legend()
```

```python
# user defined function
x = np.linspace(-10,10,1000)

def f(x):
    y = x**2 + 2*x + 2
    return y
```

```python
y_fun = f(x)
```

```python
# finding some points to plot from the function

np.column_stack((x,y_fun))[600], np.column_stack((x,y_fun))[700]
```

```python
# example of x & y pairs for getting the slope

d = delta_method()

d.fit_slope(x2=4.01, y2=26.14, x1=2.01, y1=10.07)
```

```python
# plotting the user defined function in context to the x & y pairs

d.plot(y_fun=y_fun)
```
