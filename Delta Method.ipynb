# Purpose:
Compute the slope at a point on a function using the delta method.  Plot the user defined function and get the slop of the point.  The compare the output of the slope to getting the derivative.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class delta_method():
    '''The delta method is used to approximate the slope of a points on a function's curve.
    This method will be used to find the slope of two points that are extrememly close to each other.
    
    Fit the model with two x & y pairs of your choosing using the fit_slope method.
    Plot your function along with the slope of the fitted points.
    '''
    
    def __init__(self):
        self.x = np.linspace(-10,10,1000) # change this if you would like
        
    def fit_slope(self,x2, y2, x1, y1):
        '''fit_slope method:
        computes the slope between two different x & y values
        These two points should be extremely close.
        For example:
            (2,10) & (2.0000001,10.000000600000009)
            
        Returns: slope
        '''
        self.x2 = x2
        self.x1 = x1
        self.y2 = y2
        self.y1 = y1
        self.slope = (self.y2-self.y1)/(self.x2-self.x1)
        return self.slope
    
    def plot(self, y_fun):
        '''plot method:
        y_fun: the y values from your self created function.
        The x values used in your function should be the same as
        the x values initialized here, x = np.linspace(-10,10,1000).
        You can change those x values in the code to suit your needs.
        
        Returns: the plots of your function and the slope's line at the points. 
        '''
        b = self.y2 - self.slope*self.x2 # the y intercept with respect to the slope
        y3 = self.slope*self.x + b # computing the y values with respect to the intercept
        plt.plot(self.x,y3, label=f'Slope {self.slope:.2f}') # plotting the line of the slope
        
        plt.plot(self.x, y_fun, label='Function') # plotting the function's curve
        plt.scatter(self.x2,self.y2, c='orange') # plotting the points
        plt.scatter(self.x1,self.y1, c='green') # plotting the points
        # these points should overlap since they are basically the same
        plt.legend()
```

```python
# user defined function

x = np.linspace(-10,10,1000) # x values to create the function's line plot

def f(x):
    y = x**2 + 2*x + 2
    return y
```

```python
# this will be used to plot the main curve of the function and all the different 
# points of the x value created with np.linspace(-10,10,1000)
y_fun = f(x)
```

```python
# getting the y value when x = 2
# this is to plot a point to get the slope at said point
# (2, 10)
f(2)
```

```python
# getting the y value when x = 2.0000001
# this will be the delta change (0.0000001) in the x value
# this point is meant to be extremely close to the other point above
# (2.0000001, 10.000000600000009)
f(2.0000001)
```

```python
# example of x & y pairs for getting the slope

d = delta_method()

d.fit_slope(x2=2.0000001, y2=10.000000600000009, x1=2, y1=10)
# the same points we computed above
# slope is ~ 6
```

```python
# plotting the user defined function in context to the x & y pairs and the slope

d.plot(y_fun=y_fun)
```

```python
# this method shows us that the slope for (2,10) is 6

# now lets compute the derivative of the function to see if the results 
# agree with the delta method.
```
