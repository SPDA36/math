This notebook is for demonstrating system of equations and verifying the results by ploting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-20,20,100) # this will be used for all the plots
```

```python
# y = 3x
# -5x + 2y = 2

# -5x + 2(3x) = 2
# -5x + 6x = 2
# x = 2

# y = 3(2) = 6

# -5(2) + 2(6) = 2
# -10 + 12 = 2
# 2 = 2 
```

```python
y1 = 3*x
y2 = (2+(5*x))/2
```

```python
plt.plot(x,y1, label='y1')
plt.plot(x,y2, label='y2')
plt.legend()
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.axhline(6, c='r', ls='--')
plt.axvline(2, c='r', ls='--')
plt.xlabel('x')
plt.ylabel('y')
```
