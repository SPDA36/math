This notebook is for demonstrating system of equations and verifying the results by ploting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = 3x
# -5x + 2y = 2

# -5x + 2(3x) = 2
# -5x + 6x = 2
# x = 2

# y = 3(2) = 6

# -5(2) + 2(6) = 2
# -10 + 12 = 2
# 2 = 2 
```

```python
y1 = np.linspace(0,100,1000)

x2 = y1/3
```

```python
x3 = (2 - 2*y1)/-5
```

```python
plt.plot(x2,y1, label='x2')
plt.plot(x3,y1, label='x3')
plt.legend()
plt.xlim(1,3)
plt.ylim(5,7)
plt.axhline(6, c='purple', ls='--')
plt.axvline(2, c='purple', ls='--')
```

```python
# x+y = 6
# y = 6-x

# 2x+3y = 16

# 2x + 3(6-x) = 16
# 2x + 18 - 3x = 16
# -x + 18 = 16
# -x = -2
##### x = 2

# y = 6-(2)
##### y = 4

# 2(2) + 3(4) = 16
# 4 + 12 = 16
```

```python
y = np.linspace(0,100,1000)

x1 = 6-y
x2 = (16-(3*y))/2
```

```python
plt.plot(x1,y, label='x1')
plt.plot(x2,y, label='x2')
plt.legend()
plt.xlim([1,3])
plt.ylim([3,5])
plt.axhline(4, c='purple', ls='--')
plt.axvline(2, c='purple', ls='--')
```
